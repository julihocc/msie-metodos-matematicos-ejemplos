```python
import scipy.optimize as opt

# Define the function
def f(x):
    return x**3 - 6*x**2 + 11*x - 6

# Find the root in the interval [2, 4]
root = opt.brentq(f, 2, 4)

print(f"The root of the function in the interval [2, 4] is: {root}")
```

```python
import scipy.integrate as integrate
import numpy as np

# Define the function
def f(x):
    return np.exp(-x**2)

# Compute the integral in the interval [0, 1]
result, error = integrate.quad(f, 0, 1)

print(f"The integral of the function in the interval [0, 1] is: {result}")
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the differential equation
def dydx(x, y):
    return y - x**2 + 1

# Initial condition
y0 = [0.5]

# Interval of integration
x_span = (0, 5)

# Solve the differential equation
solution = solve_ivp(dydx, x_span, y0, t_eval=np.linspace(0, 5, 100))

# Plot the solution
plt.plot(solution.t, solution.y[0])
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solution of the differential equation y' = y - x^2 + 1")
plt.grid(True)
plt.show()
```
