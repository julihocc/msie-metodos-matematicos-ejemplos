```python
# Step 1: Define variables and functions
t,a,b = var('t, a, b')
x = function('x')(t)
y = function('y')(t)

assume(t, 'real')
# Step 2: Write down the system of equations
eq1 = diff(x, t) == y
eq2 = diff(y, t) == -x

# Step 3: Solve the system of differential equations
solution = desolve_system([eq1, eq2], [x, y], ics=[0 ,a,b])

# Display the solution
solution
```
